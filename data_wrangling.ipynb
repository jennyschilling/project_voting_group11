```python
import pandas as pd
import numpy as np
```

```python
demographics_data = pd.read_csv('/content/project_voting_group11/data/va_2016-20_demographics.csv')
voting_data = pd.read_csv('/content/project_voting_group11/data/voting_VA_2020.csv')

(demographics_data.head(), voting_data.head())
```

```python
# We need to merge these two files on the county names, so we have to get the names
# formatted in the same manner first
def format_county_name(county_name):
    county_name = county_name.replace(" County", "")
    return county_name.upper()

demographics_data['COUNTY'] = demographics_data['COUNTY'].apply(format_county_name)
demographics_data.head()
```

```python
# Now we can merge the two datasets on the county name
merged_df = pd.merge(voting_data, demographics_data, left_on='county_name', right_on='COUNTY')
merged_df.to_csv('./project_voting_group11/data/merged.csv', index=False)
merged_df.head()
```

```python
for column in merged_df.columns:
  print("column = ", column)
  print("type = ", merged_df[column].dtype)
  print(merged_df[column].unique())
  print("----------------------")
  print("----------------------")
```

```python
#Renaming coded columns
new_column_names = {"AM0KE001" :    "Total",
        "AM0KE002" :   "Born in US",
        "AM0KE003":    "Born in US Territory",
        "AM0KE004":    "Born Abroad",
        "AM0KE005":    "Naturalized",
        "AM0KE006":    "Not US Citizen",
}
merged_df.rename(columns=new_column_names, inplace=True)
```

```python

#Most of the data is encoded, refered to codebooks
#Simplified data by adding all specific below poverty metrics to one column.
povertyCols = ["AM63E002","AM63E003","AM63E004","AM63E005","AM63E006","AM63E007","AM63E008","AM63E009","AM63E010","AM63E011","AM63E012","AM63E013","AM63E014","AM63E015","AM63E016","AM63E017","AM63E018","AM63E019","AM63E020",
                "AM63E021","AM63E022","AM63E023","AM63E024","AM63E025","AM63E026","AM63E027","AM63E028","AM63E029","AM63E030"]
merged_df['Total Pop Below Poverty'] = merged_df[povertyCols].sum(axis=1)
merged_df= merged_df.drop(columns=povertyCols, axis=1)
```

```python
merged_df.rename(columns={'AM63E001':'Total Pop'}, inplace=True)
```

```python
#Summing above poverty level data
abovePovertyCols = ["AM63E031","AM63E032","AM63E033","AM63E034","AM63E035","AM63E036","AM63E037","AM63E038","AM63E039","AM63E040","AM63E041","AM63E042","AM63E043","AM63E044","AM63E045","AM63E046","AM63E047","AM63E048","AM63E049",
                "AM63E050","AM63E051","AM63E052","AM63E053","AM63E054","AM63E055","AM63E056","AM63E057","AM63E058","AM63E059"]
merged_df['Total Pop At/Above Poverty'] = merged_df[abovePovertyCols].sum(axis=1)
merged_df = merged_df.drop(columns=abovePovertyCols, axis=1)
```

```python
#Removing already accounted for poverty by age data
PovertybyAgeCols = ['AM70E001','AM70E002',"AM70E003","AM70E004","AM70E005","AM70E006","AM70E007","AM70E008","AM70E009","AM70E010","AM70E011","AM70E012","AM70E013","AM70E014","AM70E015","AM70E016","AM70E017"]
merged_df = merged_df.drop(columns=PovertybyAgeCols, axis=1)
```

```python
#Rename Premade Population columns
merged_df.rename(columns={'AM8FE002':'Total Male Pop'}, inplace=True)
merged_df.rename(columns={'AM8FE021':'Total Female Pop'}, inplace=True)
merged_df.rename(columns={'AM8FE028':'Total Young-adult Women'}, inplace=True)
merged_df.rename(columns={'AM8FE009':'Total Young-adult Men'}, inplace=True)
merged_df.rename(columns={'AM8FE031':'Total Midage Women'}, inplace=True)
merged_df.rename(columns={'AM8FE012':'Total Midage Men'}, inplace=True)
```

```python
#Sum total disabled population
DisabiltyCols = ["AM8FE004","AM8FE007","AM8FE010","AM8FE013","AM8FE016","AM8FE019","AM8FE023","AM8FE026","AM8FE029","AM8FE032","AM8FE035","AM8FE038"]
merged_df['Total Disabled Pop'] = merged_df[DisabiltyCols].sum(axis=1)
```

```python
#Sum Total Children
ChildCols = ["AM8FE003","AM8FE006","AM8FE022","AM8FE025"]
merged_df['Total Children'] = merged_df[ChildCols].sum(axis=1)
```

```python
#Sum Total Senior Citizens
SeniorCols = ["AM8FE015","AM8FE018","AM8FE034","AM8FE037"]
merged_df['Total Seniors'] = merged_df[SeniorCols].sum(axis=1)
```

```python
#Drop all unused columns
colsToDrop= ["AM8FE001","AM8FE003","AM8FE004","AM8FE005","AM8FE006","AM8FE007","AM8FE008","AM8FE010","AM8FE011","AM8FE013","AM8FE014","AM8FE015","AM8FE016","AM8FE017","AM8FE018","AM8FE019","AM8FE020","AM8FE022","AM8FE023","AM8FE024","AM8FE025",
             "AM8FE026","AM8FE027","AM8FE029","AM8FE030","AM8FE032","AM8FE033","AM8FE034","AM8FE035","AM8FE036","AM8FE037","AM8FE038","AM8FE039", "AM9SE001"]
merged_df= merged_df.drop(columns=colsToDrop, axis=1)
```

```python
for column in merged_df.columns:
  print(column)
```

```python
partiallyCleanedDf= merged_df
partiallyCleanedDf.to_csv('./project_voting_group11/data/partiallyCleaned.csv', index=False)
```

```python
# identifiers and area codes not helpful so only keeping certain columns
Df = partiallyCleanedDf.iloc[:, [4, 8, 9, 10] + list(range(-18, 0))]
# for column in selected_columns.columns:
  # print(column)

# dropping Libertarian and other assuming they are not likely to win
Df = Df[~Df['party'].isin(['LIBERTARIAN', 'OTHER'])]
# Df.head()
```

```python
# finding net vote by county
counties = Df['county_name'].unique()
county_net_votes = []

for county in counties:
    county_data = Df[Df['county_name'] == county]
    total_dem = county_data[county_data['party'] == 'DEMOCRAT']['candidatevotes'].sum()
    total_rep = county_data[county_data['party'] == 'REPUBLICAN']['candidatevotes'].sum()
    # positive net vote is rep win
    net_vote = np.arcsinh(total_rep - total_dem)  # rescaled for easier prediction fitting
    county_net_votes.append(net_vote)

# only keeping first row for each county bc demographic data repeated for each mode of voting
Df = Df.drop_duplicates(subset='county_name', keep='first')
# dropping party column bc no longer needed
Df = Df.drop(columns=['party'])

# adding in net_vote as new column
Df['netvotes'] = county_net_votes

print(Df.head())
Df.to_csv('./project_voting_group11/data/cleaned_VA_voting.csv', index=False)
```
